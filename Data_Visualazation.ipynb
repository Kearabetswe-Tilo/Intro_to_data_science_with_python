```python
import numpy as np 
import pandas as pd 
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df_bank = pd.read_csv(r"C:\Users\Keara\Downloads\Intro_Data_Science_with_Python\bank.csv", delimiter = ";")
```

```python
df_bank.isnull().sum().sum()
```

```python
print(df_bank.isnull().sum())
```

```python
df_bank.describe()
```

```python
#histograms
plt.figure(figsize=(10,8))
df_bank[['age', 'nr.employed', 'cons.price.idx', 'cons.conf.idx']].hist(figsize = (10, 8), bins = 5, color = 'r')
plt.tight_layout()
plt.show()
```

```python
#bar chart
plt.figure(1)
plt.subplot(221)
df_bank['job'].value_counts(normalize = True).plot(figsize = (10, 8), kind = 'bar', color = 'red')
plt.title("Job Frequency Diagram")
plt.ylabel('Number of Customer')
plt.xlabel('Job');
```

```python
plt.subplot(222)
df_bank['marital'].value_counts(normalize = True).plot(figsize = (10, 8), kind = 'bar', color = 'green')
plt.title("Marital Status Frequency Diagram") 
plt.ylabel('Number of Customer')
plt.xlabel('Marital Status');
```

```python
plt.subplot(223)
df_bank['education'].value_counts(normalize = True).plot(figsize = (10, 8), kind = 'bar', color = 'purple')
plt.title("Education Frequency Diagram")
plt.ylabel('Number of Customer')
plt.xlabel('Education');
```

```python
plt.subplot(224)
df_bank['default'].value_counts(normalize = True).plot(figsize = (10, 8), kind = 'bar', color = 'orange')
plt.title("Default Frequency Diagram")
plt.ylabel('Number of Customer')
plt.xlabel('Default');
plt.tight_layout()
plt.show()
```

```python
#heatmap
corr = df_bank.corr()
plt.figure(figsize = (20, 10))
a = sb.heatmap(corr, annot = True, fmt = '.2f')
a.set_ylim(0, 10)
```

```python
#heatmap
corr = df_bank.corr()
plt.figure(figsize = (20, 12))
a = sb.heatmap(corr, annot = True, fmt = '.2f', cmap="YlGnBu")
a.set_ylim(0, 12)
```

```python
#box plot
plt.figure()
plt.rcParams['figure.figsize'] = (23, 10)
ax = sb.boxplot(x = "job", y = "duration", data = df_bank)
```

```python
#categorical plot
sb.catplot(data = df_bank, x = "education", y = "duration", hue = "marital", kind = "point", height = 5, aspect = 2)
```

```python

#factor plot
sb.catplot(kind = "point", data = df_bank, x = "default", y = "duration", col = "housing", row = "marital")
```

```python
#pairs plot
g = sb.pairplot(df_bank[["age", "duration", "nr.employed", "default", "y", "campaign"]], hue = "y", diag_kind = "hist")
```
