```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_hp = pd.read_csv(r"C:\Users\Keara\Downloads\Intro_Data_Science_with_Python\house_pricing.csv")
```

```python
#selection of some columns
sel_cols = ['LotFrontage', 'LotArea', 'YearBuilt', 'GrLivArea', 'GarageYrBlt',
            'GarageArea', 'SalePrice']

df = df_hp[sel_cols]
df = df.loc[0:10]
```

```python
x = df['GrLivArea'].values #get column Live Area
y = df['SalePrice'].values #get column Sale Price

x = x.reshape(-1, 1)
#y = y.reshape(-1, 1)


#random forest regression
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 20, random_state = 0)
regressor.fit(x, y)

y_pred = regressor.predict(x)
```

```python
#sort x
s_x = np.sort(x, axis = None).reshape(-1, 1)

plt.figure()
x_grid = np.arange(min(s_x), max(s_x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid), color = 'blue')
plt.title('Live Area vs Sale Price')
plt.xlabel('Live Area')
plt.ylabel('Sale Price')
plt.show()
```

```python
#finding error
msqe = sum((y_pred - y) * (y_pred - y)) / y.shape[0]
rmse = np.sqrt(msqe)
print(rmse)
```

```python
df_hp = pd.read_csv(r"C:\Users\Keara\Downloads\Intro_Data_Science_with_Python\house_pricing2.csv")
```

```python
#selection of some columns
sel_cols = ['LotFrontage', 'LotArea', 'YearBuilt', 'GrLivArea', 'GarageYrBlt',
            'GarageArea', 'Heating', 'SalePrice']

df = df_hp[sel_cols]

x = df.iloc[:, : -1].values
y = df.iloc[: , 7].values
```

```python
print(x)

print(df.isnull().sum())
```

```python
#encoding the categorical variable (heating)
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('encoder', OneHotEncoder(), [6])], remainder = 'passthrough')
x = np.array(ct.fit_transform(x))

x = x.astype(float)
```

```python
#avoid dummy variable duplications
x = x[:, 1:]

#imputation for lotFrontage and garageYrBlt
from sklearn.impute import SimpleImputer
missingvalues = SimpleImputer(missing_values = np.nan, strategy = 'median', verbose = 0)
missingvalues = missingvalues.fit(x[:, 1:])
x[:, 1:] = missingvalues.transform(x[:, 1:])
```

```python
#split dataset into train and test splits
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
#random forest regression
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor.fit(x_train, y_train)

y_pred = regressor.predict(x_test)
```

```python
#finding error
msqe = sum((y_pred - y_test) * (y_pred - y_test)) / y_test.shape[0]
print(msqe)
rmse = np.sqrt(msqe)
print(rmse)
mae = sum(abs(y_pred - y_test)) / y_test.shape[0]
print(mae)
```

## Additional Parameters

```python
#max depth versus error
md = 20;
md_errors = np.zeros(md)

from sklearn.ensemble import RandomForestRegressor
for i in range(1, md+1):
    regressor = RandomForestRegressor(n_estimators = 100, max_depth = i, random_state = 0)
    regressor.fit(x_train, y_train)
    y_pred = regressor.predict(x_test)
    #finding error
    msqe = sum((y_pred - y_test) * (y_pred - y_test)) / y_test.shape[0]
    md_errors[i-1] = np.sqrt(msqe)
```

```python
plt.scatter(range(1, md+1), md_errors, color = 'red')
plt.plot(range(1, md+1), md_errors, color = 'blue')
plt.xlabel('max depth')
plt.ylabel('rmse')
plt.show()
```

```python
#more hyperparameters
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, max_depth = 10, max_features = 0.5, min_samples_split = 5, random_state = 0)
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
msqe = sum((y_pred - y_test) * (y_pred - y_test)) / y_test.shape[0]
rmse = np.sqrt(msqe)
print(msqe)
print(rmse)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
rdg = Ridge(alpha = 1.5)
rdg.fit(x_train, y_train)

y_rdg_pred = rdg.predict(x_test)

rdg_msqe = sum((y_rdg_pred - y_test) * (y_rdg_pred - y_test)) / y_test.shape[0]
rdg_rmse = np.sqrt(rdg_msqe)
rdg_mae = sum(abs(y_rdg_pred - y_test)) / y_test.shape[0]
print(rdg_msqe)
print(rdg_rmse)
print(rdg_mae)
```
