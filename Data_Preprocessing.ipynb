```python
#replace ? with nan
df = df.replace('?', np.nan) 
print(df.isnull().sum().sum())
```

```python
#print column names with nan
print(df.isnull().sum())
print([col for col in df.columns if df[col].isnull().any()])
```

```python
#REMOVING NAN
#normalized losses
print(df['normalized-losses'])

df_temp = df[df['normalized-losses'].notnull()]
print(df_temp)
print(df_temp.isnull().sum())
```

```python
#IMPUTATION WITH MEAN
#print column types
print(df.dtypes)
```

```python
#mean of normalized losses
print(df_temp['normalized-losses'].astype(int).mean())

col_mean = df_temp['normalized-losses'].astype(int).mean()
```

```python
#impute mean
df['normalized-losses'] = df['normalized-losses'].replace(np.nan, col_mean).astype(int)
```

```python
#IMPUTATION WITH MEDIAN
#peak rpm
print(df['peak-rpm'])

df_temp = df[df['peak-rpm'].notnull()]
print(df_temp)
print(df_temp.isnull().sum())
```

```python
#print column types
df.dtypes
```

```python
#median of peak-rpm
print(df_temp['peak-rpm'].astype(int).median())

col_med = df_temp['peak-rpm'].astype(int).median()

#impute median
df['peak-rpms'] = df['peak-rpm'].replace(np.nan, col_med).astype(int)
```

```python
#bore
print(df['bore'])

df_temp = df[df['bore'].notnull()]
print(df_temp)
print(df_temp.isnull().sum())
```

```python
#print column types
print(df.dtypes)

#mean of bore
print(df_temp['bore'].astype(float).mean())

col_mean = df_temp['bore'].astype(float).mean()

#impute mean
df['bore'] = df['bore'].replace(np.nan, col_mean).astype(float)
```

```python
#IMPUTATION WITH MOST FREQUENT
#stroke
print(df['stroke'])
```

```python
df_temp = df[df['stroke'].notnull()]
print(df_temp)
print(df_temp.isnull().sum())
```

```python
#print column types
print(df.dtypes)
```

```python
#most frequent of stroke
print(df_temp['stroke'].value_counts())
print(df_temp['stroke'].value_counts().index[0])

most_frequent = df_temp['stroke'].value_counts().index[0]
```

```python
#impute most frequent
df['stroke'] = df['stroke'].replace(np.nan, most_frequent).astype(float)
```

```python
#num-of-doors
print(df['num-of-doors'])
```

```python
df_temp = df[df['num-of-doors'].notnull()]
print(df_temp['num-of-doors'].value_counts())
```

```python
#impute most frequent
df['num-of-doors'] = df['num-of-doors'].fillna(df_temp['num-of-doors'].value_counts().index[0])

print(df.describe())
```
