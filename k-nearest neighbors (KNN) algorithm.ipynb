```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df_heart = pd.read_csv(r"C:\Users\Keara\Downloads\Intro_Data_Science_with_Python\heart.csv")
```

```python
x = df_heart.iloc[:, : -1]
y = df_heart.iloc[:, -1]
```

```python
#number of yes-no
counter = Counter(y)
print(counter)
```

```python
#select two features
x = df_heart[['age', 'chol']] 
```

```python
#split dataset into train and test splits
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
#feature scaling - based on eucledian distance
from sklearn.preprocessing import StandardScaler
scx = StandardScaler()
x_train = scx.fit_transform(x_train)
x_test = scx.fit_transform(x_test)
```

```python
#fit kNN model to training set
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 11, metric = 'minkowski', p = 2) 
knn.fit(x_train, y_train)
```

```python
#predict the results on test set
y_pred = knn.predict(x_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
#Visualize train set results
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, knn.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
                                 alpha = 0.75, cmap  = ListedColormap(('blue', 'orange')))
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('blue', 'orange'))(i), label = j, edgecolors='black')
plt.title('k-NN (Training Set)')
plt.xlabel('Age')
plt.ylabel('Cholestrol')
plt.legend()
plt.show()
```

```python

#Visualize test set results
from matplotlib.colors import ListedColormap
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, knn.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
                                 alpha = 0.75, cmap  = ListedColormap(('blue', 'orange')))
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('blue', 'orange'))(i), label = j, edgecolors='black')
plt.title('k-NN (Test Set)')
plt.xlabel('Age')
plt.ylabel('Cholestrol')
plt.legend()
plt.show()
```

```python
x = df_heart.iloc[:, : -1]
y = df_heart.iloc[:, -1]
```

```python
#split dataset into train and test splits
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
#fit kNN model to training set
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2) 
knn.fit(x_train, y_train)
```

```python
#predict the results on test set
y_pred = knn.predict(x_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
#number of neighbors versus accuracy
accuracy = np.zeros(5)
j = 0

for i in range(1, 11, 2):
   knn = KNeighborsClassifier(n_neighbors = i, metric = 'minkowski', p = 2) 
   knn.fit(x_train, y_train)
   y_pred = knn.predict(x_test)
   #finding error
   accuracy[j] = sum(y_pred == y_test) / y_test.shape[0]
   j = j + 1
```

```python
plt.scatter(range(1, 11, 2), accuracy, color = 'red')
plt.plot(range(1, 11, 2), accuracy, color = 'blue')
plt.xlabel('k')
plt.ylabel('accuracy')
plt.show()
```
